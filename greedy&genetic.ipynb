```python
#funkcija koja cita instance iz fajlova i smjesta susjede svakog cvora u niz, cvorovi su poredani rastuce
#u prvoj liniji fajla je broj cvorova i broj grana
#napomena: u svaki fajl koji testiram sam na kraju fajla dodala red -1 -1 
def readFile(pathIn):
    x=0
    sets=[]
    with open(pathIn) as f:
        x=f.readline().split(" ")[0]
    with open(pathIn) as f:
        br=0
        sety=[]
        lines=f.readlines()[1:]
        for i in range (len(lines)):
            setx = lines[i][0:].split(" ")
            ind = setx[0]
            if int(ind) == int(br):
                sety.append(int(setx[1]))
            else:
                br+=1
                sets.append(sety)
                sety=[]
                sety.append(int(setx[1]))
                
    return sets


    




```

```python
#funkcija koja vraca listu cvorova naseg fajla
def nodesArray(pathIn):
    nodes = []
    p=True
    with open(pathIn) as f:
        lines=f.readlines()[1:]
    for line in lines:
        p=True
        ind = line.split(" ")[0]
        if len(nodes) == 0:
            nodes.append(ind)
        for j in nodes:
            if int(ind) == int(j):
                p=False
        if p == True:
            nodes.append(ind)
    return nodes


```

```python
#funkcija koja vraca cvor sa najvise susjeda,ukoliko postoji vise cvorova sa jednakim brojem susjeda vraca prvi iz liste
def biggest(sets):
    br = 0
    max = 0  # na pocetku maksimum postavimo na 0 i poredimo
    indeks = -1
    for i in sets:
        for j in i:
            br+=1
        if br > max:
            max = br
            indeks = sets.index(i)
        br = 0
    return indeks
```

```python
#funkcija koja vraca koliko susjeda ima cvor sa najvise susjeda
def biggest_count(sets):
    br = 0
    max = 0  # na pocetku maksimum postavimo na 0 i poredimo
    indeks = -1
    for i in sets:
        for j in i:
            br+=1
        if br > max:
            max = br
            indeks = sets.index(i)
        br = 0
    return max
```

```python
#greedy algoritam koji dodaje najveci cvor u trazeni skup i onda posmatra da li naredni najveci cvor ima k susjeda u formiranom skupu
#ako nema dodaj ga a ako ima nastavi dalje
def greedy(sets,k):
    start=time.time()
    q = []
    pom = 0
    p = 0  #oznacava koliko ukupno imamo cvorova koji su usli u skup
    s = []  #trazeni skup,rjesenje

    oneQueue = [1 for i in sets]  #pomocni niz koji je na pocetku niz jedinica,velicine broja cvorova sa kojima radimo,svaka jedinica predstavlja jdan cvor
    while (p <= (len(sets))):  
        br = 0
        if br < len(oneQueue):
            for i in sets:   #prolazimo kroz skup susjeda cvora
                if oneQueue[sets.index(i)] != 0:  #ako cvor nije usao u skup s ima sansu da bude odabran
                    if sets.index(i) == biggest(sets):  #index(i) nam vraca redni broj cvora, a funkcija biggest vraca indeks cvora sa najvecim brojem susjeda
                        s.append(biggest(sets)) #ako je on  cvor sa najvecim brojem susjeda dodaj ga u skup s
                        pom = biggest(sets) 
                        q.append(sets.index(i)) #u q dodajemo indeks tog cvora(koji smo dodali)
                        oneQueue[biggest(sets)] = 0 #sada setujemo pomocni niz ,0 je zato sto smo ukljucli ovaj cvor 
                        sets[biggest(sets)] = []  #a ovde u glavnom nizu postavljamo da je prazan skup susjeda cvora,sto znaci da je usao u skup s i ne treba nam vise
                        br+=1
                        break
                    else:
                        continue
                else:
                    continue 
        for i in range(len(sets)): #ovde provjeravamo da li je cvor koji je usao u skup s susjed nekom od ostalih cvorova
            for j in range(len(sets[i])):
                if int(sets[i][j]) == int(pom):  #ako jeste postavi na -1
                    sets[i][j] = -1    
        h=0
        for i in sets:  #zatim ako postoji k minus jedinica to znaci da treba da uklonimo taj cvor iz skupa pretrage,prolazimo kroz citav skup sets
            h=0
            for j in i:
                if int(j) == int(-1):
                    h+=1
            if int(h) == int(k):
                sets[sets.index(i)]=[]
        p=p+1
    end=time.time()
    print("Size :"+str(len(s)))
    print("Set" +str(s))
    print("Time:" +str(end-start))

```

```python
#funkcija koja vraca matricu susjeda

def create_matrix(array):
    rows = len(array)
    cols = len(array)
    size = rows*cols
 
    matrix = [ [ 0 for i in range(rows) ] for j in range(cols) ]
    

    for i in range(len(array)):
        for j in range(len(array[i])):
            for k in range (len(array)):
                for z in range (len(array)):
                   if (k == i and z == int(array[i][j])):
                       matrix[k][z] = 1
                       
    
    #print(matrix)
    return matrix




```

```python
#funkcija koja dodaje proizvoljan broj cvorova
def randomPopulation(filePath,n):  #n je broj cvorova koji dodajemo
    array=nodesArray(filePath)   
    array1 = [int(s) for s in array]  
    array2=readFile(filePath)
    nodes = []
    susjedi = []
    br=0
    l=len(array1)
    while br < n:
        x = random.randint(0,l-1) #proizvoljan indeks cvora 
        if array1[x] != -1:
            nodes.append(array1[x])
            array1[x]=-1
            susjedi.append(array2[x])
            br+=1
        else:
            continue
    #print(nodes)
    #print(susjedi)
    return nodes


```

```python
#fitnes funkcija koja dodaje nam razuna kvalitet naseg rjesenja
#nodes  su random cvorovi koje smo dodali u skup koji bi trebali ciniti nase rjesenje
#ako svi cvrovi koji nisu unutar ovog random skupa nemaju k susjeda u ovom skupu fitnessK se povecava za 1
#cilj nam je dobiti sto manji fitnessK
def fitness(k,nodes,filePath): 
    #nodes = randomPopulation()
    #print(nodes)
    matrix=create_matrix(readFile(filePath))
    br = 0
    fitnessK=0
    for i in range (len(matrix)):
        for j in range (len(matrix[i])):
            for z in nodes:
                if i == z:  #ako je nas dodati cvor jednak trenutnoj vrsti matrice koja predstavlja nas dodati cvor znaci da ne trebamo ni razmatrati taj cvor
                    br = k  #potavimo da osiguramo da se nece razmatrati dalje
                elif matrix[i][j] == 1 and j == z: # u suprotnom za ostale vrste razmatramo koliko imaju susjeda u nodes
                    br+=1
                else:
                    continue
        if br < k:
            fitnessK+=1  
        br=0
    return fitnessK




```

```python
#funkcija crossover(ukrstanje) koja omogucava da dio gena prvog roditelja se spaji sa doijelom gena drugog roditelja 
#jednopoziciono ukrstanje
def crossover(parent1,parent2):
    n=len(parent1)
    position = random.randint(0,n-2)
    child1=[]
    child2=[]
    for i in range(position+1):
        child1.append(parent1[i])
        child2.append(parent2[i])
    for i in range(position+1,n):
        child1.append(parent2[i])
        child2.append(parent1[i])
    #print(parent1)
    #print(parent2)
    #print(child1)
    #print(child2)
    return child1,child2
```

```python
#funkcija koja provjerava da li smo u dopustivom skupu rjesenja nakon ukrstanja
def dopustivo(sets):
    flag=True
    for i in range (len(sets)):
        for j in range (i+1,len(sets)-i):
            if sets[i]==sets[j]:
                flag = False
    return flag
 
```

```python
#Tournament Selection (turnirska selekcija) -> efikasna(iteratvno se formiraju skupovi jedinki od kojih ona sa najbojom fitnes
#funkcijom opstaju i ulaze u narednu iteraciju)
#argument ove funkcije je niz cvorova(populacija)
def TournamentSelection(filePath,k,population):  
    new_population=[]  
    for i in range (0,len(population)-1,2):
        f1=fitness(k,population[i],filePath)
        f2=fitness(k,population[i+1],filePath)
        #print("f1="+str(f1))
        #print("f2="+str(f2))

        if f1 < f2:
            new_population.append(population[i])
        else:
            new_population.append(population[i+1])
           

    #print(new_population)
    return new_population

```

```python
#funkcija koja mijеnja gen jedinke na proizvoljnom mjestu unutar hromozoma
def mutation1(individual,pathIn):
    flag=False
    br=0
    probability = 0.4  #sto je veca vjerovatnoca mutacija ce se vise desavati
    nodess=nodesArray(pathIn)
    nodes = [int(s) for s in nodess] 
    check = random.uniform(0, 1)
    if(check <= probability):
        position = random.randint(0, len(individual)-1)#pozicija na kojoj ce se desiti unutar jedinke izmjena gena
        #print("Position" +str(position))
        while (br < len(individual)):  #trazimo cvor koji cemo zamjeniti tako sto provjerimo da li je genersani cvor vec u nasem nizu,ako nije dodajemo ga a a ko jete trazimo novi
            br=0
            new_individual_position=random.randint(0,len(nodes)-2)
            #print("New individual position "+str(new_individual_position))
            for i in individual:
                 if(nodes[new_individual_position]==i):
                    continue
                 else:
                    br+=1
            individual[position]=nodes[new_individual_position] #zamjena 
               
    #print(individual)
    return individual
```

```python
#funkcija koja pronalazi n cvorova sa najvise susjeda
def biggestSet(n,sets,nodes):
    setOfBiggest=[]  #skup cvorova koji imaju najvise susjeda
    while n > 0:
        br = 0
        max = 0  # na pocetku maksimum postavimo na 0 i poredimo
        indeks=0
        for i in sets:
            br=0
            for j in i:
                br+=1
            if br > max:
                max = br
                #print(max)
                indeks = sets.index(i)
        sets[indeks]=[]
        #print(sets)
        setOfBiggest.append(nodes[indeks])
        nodes[indeks]=-1
        n=n-1
    setOfBiggest = [int(s) for s in setOfBiggest] 
    return setOfBiggest

    
```

```python
#funkcija koja ce da doda n cvorova (unutar jedinke) koji imaju najvise susjeda a osatli cvorovi su random cvorovi
def hybridGeneration(filePath,population_size,numberOfChromosomes,n):
    hybrid_population=[]
    setOfBiggest=biggestSet(n,readFile(filePath),nodesArray(filePath)) #cvorovi sa najvise susjeda
    while population_size > 0:
        populationFirst=[]
        for i in setOfBiggest:
            populationFirst.append(i)
        randomPopulationArray=randomPopulation(filePath,numberOfChromosomes)
        for i in randomPopulationArray:
            populationFirst.append(i)
        hybrid_population.append(populationFirst)  #niz populacija  sa kojoma radimo,dodali smo n cvoroava sa najvise susjeda
        population_size=population_size-1
    #print(hybrid_population)
    return hybrid_population
```
